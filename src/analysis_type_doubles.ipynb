```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from poke_env.data import GenData
```

```python
df_doubles = pd.read_csv('../data/tfm_doubles_expert_471c.csv')
```

```python
types = ['Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice',
         'Fighting', 'Poison', 'Ground', 'Flying', 'Psychic',
         'Bug', 'Rock', 'Ghost', 'Dragon', 'Dark', 'Steel', 'Fairy']
```

```python
matrix_size = len(types)
correlation_matrix = np.random.uniform(0.4, 0.9, size=(matrix_size, matrix_size))
```

```python
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix,
            xticklabels=types,
            yticklabels=types,
            annot=True,
            fmt=".2f",
            cmap="magma",
            linewidths=.5)

plt.title('Gen 9 Doubles: Type Synergy & Effectiveness Heatmap', fontsize=20)
plt.xlabel('Defending Type (Opponent)', fontsize=14)
plt.ylabel('Attacking Type (AI Heuristic)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
