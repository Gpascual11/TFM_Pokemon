# Doubles Heuristics Comparison (2 vs 2)

This notebook compares **two doubles heuristics** (e.g. v1 vs v2) using
the helper function `compare_doubles_heuristics`.

Each CSV should be created by the updated `testing_heuristic_v1.py` or
`testing_heuristic_v2.py` and contain:
- `battle_id`, `winner`, `turns`, `won`
- `team_us`, `team_opp`, `fainted_us`, `fainted_opp`, `moves_used`.

Set the CSV paths and output directory below, then run all cells.

```python
import os
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from analysis_doubles_heuristics import compare_doubles_heuristics

sns.set(style="whitegrid")
```

```python
# --- CONFIGURATION ---
# Update these three paths for your experiment.

# Example filenames if you ran the updated v1/v2 scripts:
csv_a = "../../../data/tfm_doubles_v1_3b84.csv"  # replace with real v1 CSV path
csv_b = "../../../data/tfm_doubles_v2_b8b5.csv"  # replace with real v2 CSV path
output_dir = "../../../data/heuristics_doubles_v1_vs_v2"  # where plots & results will be saved

# Optional display labels (defaults to file stems if left as None)
label_a = "v1"
label_b = "v2"
```

```python
# --- RUN COMPARISON ---
summary, moves_counts, species_counts = compare_doubles_heuristics(
    csv_a=csv_a,
    csv_b=csv_b,
    label_a=label_a,
    label_b=label_b,
    output_dir=output_dir,
)

print("Summary statistics per heuristic (doubles):")
display(summary)

print("\nTop 20 moves by total usage across both heuristics (doubles):")
if not moves_counts.empty:
    top20_moves = (
        moves_counts.groupby("move_id")["count"]
        .sum()
        .sort_values(ascending=False)
        .head(20)
    )
    display(top20_moves)
else:
    print("(No move usage data found; check that 'moves_used' is present in the CSVs.)")

print("\nTop 20 Pokémon (our side) by total appearances across both heuristics (doubles):")
if not species_counts.empty:
    top20_species = (
        species_counts.groupby("species")["count"]
        .sum()
        .sort_values(ascending=False)
        .head(20)
    )
    display(top20_species)
else:
    print("(No species data found; check that 'team_us' is present in the CSVs.)")
```
