# Singles Heuristics Comparison (1 vs 1)

This notebook compares **two singles heuristic runs** using the helper
functions defined in `analysis_singles_heuristics.py` in this folder.

Each CSV should be generated by your 1v1 scripts (e.g. `test_heuristic_v5.py`)
and contain at least:
- `battle_id`, `winner`, `turns`, `won`
- `team_us`, `team_opp`, `fainted_us`, `fainted_opp`, `moves_used`.

Set the CSV paths and output directory below, then run all cells.

```python
import os
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from analysis_singles_heuristics import compare_singles_heuristics

sns.set(style="whitegrid")
```

```python
# --- CONFIGURATION ---
# Update these three paths for your experiment.

# Example: paths relative to this notebook
csv_a = "../../../../data/tfm_expert_singles_v4_4eb9.csv"  # your existing file
csv_b = "../../../../data/tfm_expert_singles_v5_ac6f.csv"  # same for now
output_dir = "../../../../data/heuristics_compare_single_run"
label_a = "v4 (run A)"
label_b = "v5 (run B)"
```

```python
# --- RUN COMPARISON ---
summary, moves_counts = compare_singles_heuristics(
    csv_a=csv_a,
    csv_b=csv_b,
    label_a=label_a,
    label_b=label_b,
    output_dir=output_dir,
)

print("Summary statistics per heuristic:")
display(summary)

print("\nTop 20 moves by total usage across both heuristics:")
if not moves_counts.empty:
    top20 = (
        moves_counts.groupby("move_id")["count"]
        .sum()
        .sort_values(ascending=False)
        .head(20)
    )
    display(top20)
else:
    print("(No move usage data found; check that 'moves_used' is present in the CSVs.)")
```
